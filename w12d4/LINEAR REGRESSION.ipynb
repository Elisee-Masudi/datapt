```python
# Cov(X, Y) = moyenne ( écart entre X et sa moyenne * écart entre Y et sa moyenne )
```

```python
# Covariance élevée : lorsque X prend des grandes valeurs "en même temps" que Y, et inversement, X prend des petites valeurs "en même temps" que Y
# Covariance faible : lorsque X prend des grandes valeurs "en même temps" que Y prend des petites valeurs, et inversement
```

```python
# Corrélation est entre -1 et 1
# Covariance est toujours inférieure au produit des écarts types de X et Y
# Corrélation = Covariance / (écart type de X * écart type de Y)
```

```python
# Corrélation = version normalisée de la covariance
```

```python
# Covariance : ça se mesure en unité de X * unité de Y
# Corrélartion : pas d'unité
```
